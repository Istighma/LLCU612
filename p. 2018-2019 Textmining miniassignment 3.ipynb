    
<h2> Word Counts of a Corpus in Python <h2>    

Herein, the goal is to count words within a text and attempt to offer an interpretation.
    In order to do so, I think the best option is to organize the text in a data structure called "dict". 
    This type of data structure allows to associate values to keys. With some basic formulations, it is 
    to split every line of a text and split every word of a line into an individual unit. This unit may 
    become a "key" of the dictionary. Once it is a key, the occurrence of a word can have a value 
    associated to it. Our goal, thus, is to give a value to all the words in the corpus we want to study. 
    Once this is completed, it will be possible to print the dictionary and grasp the frequency of words 
    our corpus.
    
In order to understand the use of dictionary, I had to do lots of research to try to understand how python dictionary work. In the end, I found this link and with the help of the python documentation and with the help of some classmates, I was able to get an outcome.
I used the following link as a model to write my program:
https://stackoverflow.com/questions/21107505/word-count-from-a-txt-file-program

```python
import os

path = "/home/otzy/Notebookana/austen/" #determine a path
filenames = os.listdir(path) #define filenames as the list of files in the path
print("The following table is a word count for the following files:" , filenames) #print the list of files

dictionary = {} #this creates a dictionary type data structure

for filename in filenames: # this starts a loop that indivuates filenames into single files
    with open(path+filename, 'r') as f: # this is a wonderful option that opens and closes a document for you.
        inside = f.read().splitlines() #this defines inside as reading f and individuating it's lines

    for line in inside:       #this starts a loop of actions that will be repeated for every line inside the file
        line = line.lower()   #this makes all lines lowercase, and because of the loop, all of the text becomes lower case
        words = line.split()  #this splits the lines into smaller parts

        for word in words:    #Another loop. this time it commands "for every word"
            if word not in dictionary:    #if a word is not in the dictionary
                dictionary[word] = 0      #integrate the new word to the dictionary as a new key
            
            dictionary[word] = dictionary[word] + 1   #for every word add a value of +1 to the key a word belongs to
            
print(dictionary) #print dictionary
```

<h3>Cleaning the results

This is a workable outcome, but not a desirable one. I see two problems. First, it is hard to read to the human eye. Second, it is not classified by descending values like it is in Voyant.

With further research, it was possible to solve these problems.

To do so, I did some more research on how to clean data. I found this link to be satisfying: https://stackoverflow.com/questions/20577840/python-dictionary-sorting-in-descending-order-based-on-values

```python
dirty = dictionary
dirty_cleaned = sorted(dirty, key=dirty.get, reverse=True)
for r in dirty_cleaned: #this says: for any key in sorted dict.
    print(r, dirty[r]) #this says: print key and value next to it.
```
