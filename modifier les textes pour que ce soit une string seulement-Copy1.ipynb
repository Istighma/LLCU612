Below, I want to read every file and find the paragraphs that contain the word temessuk. I want to find the word temessuk, look at what's behind, look at what's after.

I decided to mark any \n\n with 'marker' and then delete all remaining \n to create a single string. it makes the re.findall(x, text) more efficient in that it allows me to avoid integrating \n in my re.findall equation.

The problem I face concerns is this:
(?<= marker ).*temessük.+?(?= marker )
the (?<= marker ).* is problematic for it says: find "marker" match any character until "temessük". So it still matches the "marker" after the "marker"... so I tried writing 
        paragraphs = re.findall(r"(?<= marker.? )[^\bmarker\b].*temessük.+?[^\bmarker\b](?=marker)", text)
 But it won't work.

```python
for filename in filenames:
    with open(path+filename, 'r') as f:
        text = f.read()
        text = re.sub("\n\n", " marker ", text)
        text = re.sub("\n", "", text)
        paragraphs = regex.findall(r'(?<= marker ).*temessük.+?(?= marker )', text)
        text = "\n".join(paragraphs)
        print(text)
```
