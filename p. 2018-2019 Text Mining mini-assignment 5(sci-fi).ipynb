I want to spot words that come up often in two corpuses of sci-fi texts. I chose and imported files from 5 movie stripts of the 1990s, and 5 movie scripts from the 2010s. Although chosen arbitrarily, the following program can be run on an infinite number of local scripts.

the files I downloaded were separated in two different folders manually (through the inferace). I thus prepare corpus one in Corpus 1 file(movies from 1990s), and then I do the same steps for Corpus 2 (movies from the 2010s).

To do this, one needs first to import the right programs and create a workable variable with the files. printing the list of files is a good idea to know if we're at the right place.

```python
import os
import re

path = "/home/otzy/Notebookana/Corpus 1/" #this determines the folder in which the files are
filenames = os.listdir(path)              #associates the various files inside the foler to filenames
filenames                                 #prints file names as list, confirms that all files are executable
```

Movie scripts have a lot of noise that render tokenization difficult. For this reason, I try to keep the text to its simplest form by removing all punctuation and capital letters. From there, tokenization of words will be easy.

To create a corpus, I merge all scripts into a single file. This new file named 'compiled' becomes my new folder.

```python
for filename in filenames:                            #looping all files through the same steps
    with open(path+filename, 'r') as f:               #opens & closes file as readable
        string = f.read()                             #creates a workable string
        short_text = re.sub("([?]|[!]|[:]|[;]|[,]|[.])", "", string)   #relimintes punctuation
        tokenizable = short_text.lower()              #lowers the caps
        with open(path+"compiled", 'a+') as f:        #creates a new file in the folder as readable, writable, and appendable
            f.write(tokenizable)                      #for each file writes texts of scripts without punctuation and in lowercaps
```

Now that I have built this file, I can start to tokenize the words in it, count and print the counts in a table.

To do so, I import Natural Language Toolkit Kernel and other programs in order to facilitate my work.

```python
import nltk              #imports nltk (Natural Language Toolkit Kernel)
from nltk.tokenize import sent_tokenize, word_tokenize   #from nltk.tonize server import these two programs

stopwords = nltk.corpus.stopwords.words("english")       #nltk stopwords programmed to english stopwords

with open(path+"compiled", 'r') as f:  #read the file created earlier
    string = f.read()                  #creates string
    tokens1 = nltk.word_tokenize(string)   #creates token1 from a function in nltk word_tokenize program
```

```python
nowords1 = [word for word in tokens1 \
        if word[0].isalpha() and word not in stopwords]     #takes stopwords out of token list
print("This is a sample of tokenized words:\n", nowords1[:100]) #prints 100 first keys of token list
```

We can notice from this sample of tokenized words that many words are pluralized. This would create problem if I created a dictionary from these items. I thus want to remove plurals. For this reason I import WordNetLemmetizer which changes plural words into their singular forms.

```python
from nltk import WordNetLemmatizer #imports nltk

for word in nowords1:        #for each word in the no stop-words list
    word_lemmetize = WordNetLemmatizer.lemmatize(word,word) #this creates a variable with only singular words
```

```python
tokens = nowords1        #no words is too abstarct, I change it to tokens
lemmetize_tokens = [WordNetLemmatizer.lemmatize(word,word) for word in tokens]  #
retoken = [word for word in lemmetize_tokens if word[0].isalpha()] #tokens start only with [a-z]
word_frequency = nltk.FreqDist(retoken) #activates the nltk word-counting option
#this is a copy-paste of an earlier cleaning program I did
#I just change the variable, it's easier this way.
#I could eventually upload this to a package I import whenever I program with python3
dirty = word_frequency  #determine the dirt print results
dirty_cleaned = sorted(dirty, key=dirty.get, reverse=True)  #sorts the words in decreasing order
for r in dirty_cleaned: #this says: for any key in the list
    print(r, dirty[r]) #print r which means the concerned key, and dirty[r] as the value associated with it
```
