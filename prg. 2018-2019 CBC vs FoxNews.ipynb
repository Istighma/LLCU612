```python
import urllib.request
response = urllib.request.urlopen('https://www.foxnews.com')
html = response.read()
html



from bs4 import BeautifulSoup
soup = BeautifulSoup(html, 'lxml')
soup

for title in soup.find_all("h2", class_="title"):
    print(title.text)
    
for title in soup.find_all("h3", class_="title"):
    print(title.text)
```

```python
import urllib.request
response = urllib.request.urlopen('https://www.cbc.ca')
html = response.read()
html

from bs4 import BeautifulSoup
soup = BeautifulSoup(html, 'lxml')
soup

for headline in soup.find_all("h3", class_="headline"):
    print(headline.text)
```
